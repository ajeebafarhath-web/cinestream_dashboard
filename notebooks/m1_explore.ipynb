#### Data Quality Report

Issues Found:

1. Type column contains case mismatches.
2. Genre column contains trailing spaces.
3. Language column contains trailing spaces.
4. Director column contains missing values.
5. AddedDate column contains missing values.
6. IMDbScore contains 3 values outside the valid range (1–10).
7. CriticRating contains 2 values outside the valid range (1–10).
8. RuntimeMinutes contains an extreme outlier value (99999 minutes).
9. SubscribersGainedThousands contains 81 negative values.
10. Duplicate rows found.

Start Cleaning:

```python
clean_df = df.copy()
```

strip whitespace in text columns

```python
text_cols = ['Title',
    'Type',
    'Genre',
    'Language',
    'Country',
    'Director',
    'AgeRating',
    'TrendingStatus'
]
```

```python
for col in text_cols:
    clean_df[col] = clean_df[col].str.strip()
```

Standardise case in Type Column

```python
clean_df['Type'] = clean_df['Type'].str.title()
```

Drop duplicates

```python
clean_df = clean_df.drop_duplicates()
```

```python
clean_df.duplicated().sum()
```

Replace out-of-range ratings and runtimes with NaN

```python
clean_df.loc[
    (clean_df['IMDbScore'] < 1) |
    (clean_df['IMDbScore'] > 10),
    'IMDbScore'
] = np.nan
```

```python
clean_df.loc[
    (clean_df['CriticRating'] < 1) |
    (clean_df['CriticRating'] > 10),
    'CriticRating'
] = np.nan
```

```python
clean_df.loc[
    clean_df['RuntimeMinutes'] > 5000,
    'RuntimeMinutes'
] = np.nan
```

```python
clean_df['RuntimeMinutes'][clean_df['RuntimeMinutes'].isna()]
```

Replace negative subscriber counts with 0

```python
clean_df.loc[
    clean_df['SubscribersGainedThousands'] < 0,
    'SubscribersGainedThousands'
] = 0
```

Convert AddedDate to a real datetime.

```python
clean_df['AddedDate'] = pd.to_datetime(clean_df['AddedDate'],
    errors='coerce'
)
```

Create Profit_Cr Column

```python
clean_df['Profit_Cr'] = (clean_df['RevenueCr'] - clean_df['ProductionCostCr'])
```

Create ROI_Pct Column

```python
clean_df['ROI_Pct'] = (clean_df['Profit_Cr'] / clean_df['ProductionCostCr']
) * 100
```

Create Performance_Band Column

```python
clean_df['Performance_Band'] = np.where(clean_df['Profit_Cr'] > 20,'Hit',
    np.where(clean_df['Profit_Cr'] >= 0,'Break-even','Flop')
)
```

Check New Columns

```python
clean_df[['Profit_Cr','ROI_Pct','Performance_Band']].head()
```

Save Clean Dataset

```python
clean_df.to_csv('../outputs/cleaned_cinestream.csv',index=False)
```
